```python
nums=[5,2,3,4,8,9,0] # should not contain repitition
n=len(nums)
```

```python
out=[[x] for x in nums]
out
```

```python

depth=n-1
for d in range(depth):
    print('depth = ',d,' current out = :', out)
    for ele_list in out:
        print('traverse :',ele_list)
        
        last_ele=ele_list[-1]
        last_ele_idx=nums.index(last_ele)
        
        print('last element and index: ',last_ele,last_ele_idx)
        
        if(last_ele_idx+1==n):
            continue
        
        
        
        for idx in range(last_ele_idx+1,n):
            new=ele_list.copy()
            new.append(nums[idx])
            print(new)
            if(new not in out):
                out.append(new)

    print('depth = ',d,' New out = :', out)
        
    
        
        
        
```
